```python
import glob
import random
import shutil
```

```python
PATH = "C:/Users/sidha/Desktop/Tarran/Research/MVDI25K/JPEGImages/"
img_paths = glob.glob(PATH+'*.jpg')
txt_paths = glob.glob(PATH+'*.txt')
destination_folder ="C:/Users/sidha/Desktop/Tarran/Research/data/"
print(img_paths,txt_paths)
# Calculate number of files for training, validation
data_size = len(img_paths)
r = 0.7
train_size = int(data_size * 0.7)
test_size = int(data_size * 0.2)
valid_size = int(data_size * 0.1)
# Shuffle two list
img_txt = list(zip(img_paths, txt_paths))
random.seed(43)
random.shuffle(img_txt)
img_paths, txt_paths = zip(*img_txt)

# Now split them
train_img_paths = img_paths[:train_size]
train_txt_paths = txt_paths[:train_size]

test_img_paths = img_paths[train_size:test_size]
test_txt_paths = txt_paths[train_size:test_size]

valid_img_paths = img_paths[test_size:valid_size]
valid_txt_paths = txt_paths[test_size:valid_size]

# Move them to train, valid folders
train_folder_images = destination_folder+'train/images/'
train_folder_labels = destination_folder+'train/labels/' 
test_folder_images = destination_folder+'test/images/'
test_folder_labels = destination_folder+'test/labels/'
valid_folder_images = destination_folder+'valid/images/'
valid_folder_labels = destination_folder+'valid/labels/'

os.mkdir(train_folder_images)
os.mkdir(train_folder_labels)
os.mkdir(valid_folder_images)
os.mkdir(valid_folder_labels)
os.mkdir(test_folder_images)
os.mkdir(test_folder_labels)

def move(paths, folder):
    for p in paths:
        shutil.move(p, folder)

move(train_img_paths, train_folder_images)
move(train_txt_paths, train_folder_labels)
move(valid_img_paths, valid_folder_images)
move(valid_txt_paths, valid_folder_labels)
move(test_img_paths, test_folder_images)
move(test_txt_paths, test_folder_labels)
```
